# **4. Limpieza y tratamiento de los datos**

## **4.1 Data Preparation**


###**Datos Faltantes:**

Basado en el diagn贸stico anterior, se toman las siguientes decisiones estrat茅gicas para manejar los datos faltantes:

1.  **Variable `Age` (Edad):** Imputar con la **Media**. La edad es vital para determinar supervivencia (ni帽os/ancianos), y perder esos registros es muy costoso para el objetivo determinado.
2.  **Variable `Embarked` (Puerto):** Faltan solo 2 registros.
    *   *Acci贸n:* Imputar con la **Moda** (el puerto m谩s frecuente). El impacto estad铆stico es despreciable.
3.  **Variable `Cabin` (Cabina):** Faltan >70% de los datos.
    *   *Acci贸n:* **Eliminar la variable**. Intentar imputar tantos datos introducir铆a un ruido masivo y sesgo artificial en el modelo.


```python
#Limpieza de datos para train
train["Age"] = train["Age"].fillna(train["Age"].mean())

train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])


#Limpieza de datos para test
test["Age"] = test["Age"].fillna(test["Age"].mean())

test["Fare"] = test["Fare"].fillna(test["Fare"].mean())

test["Embarked"] = test["Embarked"].fillna(test["Embarked"].mode()[0])

#En ambos casos, se va a eliminar la variable "Cabin"
train = train.drop(columns=['Cabin'])
test = test.drop(columns=['Cabin'])

train.info()
```

### 3.2 Codificaci贸n de Variables (Encoding)

Los modelos de Machine Learning requieren entradas num茅ricas. Las variables categ贸ricas (`Sex`, `Embarked`) contienen informaci贸n valiosa, pero est谩n en formato texto.

**Estrategia de transformaci贸n:**
*   **Sex:** Mapeo binario directo (`male=0`, `female=1`).
*   **Embarked:** Mapeo num茅rico (`S=0`, `C=1`, `Q=2`).

**Selecci贸n de Variables (Feature Selection):**
Las variables `Name` y `Ticket` son identificadores 煤nicos de texto y alta cardinalidad. De tal manera que no aportan valor predictivo directo, por lo que se decide **excluirlas** para simplificar el modelo.

```python
for col in ["Sex", "Embarked"]:
    print(col, train[col].nunique(), train[col].unique())
```

```python
#Para train.csv
train["Sex"] = train["Sex"].map({"male": 0, "female": 1}).astype("Int64")
train["Embarked"] = train["Embarked"].map({"S": 0, "C": 1, "Q":2}).astype("Int64")

#Para test.csv
test["Sex"] = test["Sex"].map({"male": 0, "female": 1}).astype("Int64")
test["Embarked"] = test["Embarked"].map({"S": 0, "C": 1, "Q":2}).astype("Int64")

train = train.drop(columns=['Name', 'Ticket'])
test = test.drop(columns=['Name', 'Ticket'])

train.info()
test.info()
```
